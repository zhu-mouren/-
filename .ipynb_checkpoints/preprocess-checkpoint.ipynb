```python
import time
import glob
import os

import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.models import load_model
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import (
    Input,
    Conv1D,
    MaxPooling1D,
    Dropout,
    BatchNormalization,
    Activation,
    Add,
    Dense,
    GlobalAveragePooling1D,
)
import keras.backend as K
from easydl import clear_output
from sklearn.utils import shuffle
from keras.callbacks import ModelCheckpoint, EarlyStopping

# from get_representations import *

def normalization(data):
    """
    only for data
    data: 2d array 
    """
    print(data.shape)
    assert len(data.shape)==2
    
    for index,line in enumerate(data):
        max_ = line.max()
        min_ = line.min()
        data[index] = (line-min_)/(max_-min_)
    return data
```

```python
dataindexs = [i for i in range(1,26)]
# dataindexs = [i for i in range(26,46,1)]


def split_train_val_test(val:list,test:list):
    """
    完成了每个数据集的读取和shuffle
    """
    def flt(x):
        return not (x in val or x in test)
    
    def load(indexs):
        ECG = np.empty((1,4096))
        LABEL = np.empty((1,2))
        for index in indexs:
            ecg = np.load('/data/drink/ECG/{}.npy'.format(index))
            label = np.load('/data/drink/label/label{}.npy'.format(index))

            ECG = np.vstack((ECG,ecg))
            LABEL = np.vstack((LABEL,label))
        
#         ECG_ = np.zeros((ECG.shape[0],4096))
#         for i in range(ECG.shape[0]):
#             ECG_[i] = scipy.signal.resample_poly(ECG[i],4096,5000)
#         res = shuffle(ECG[1:],LABEL[1:],random_state=666)
        print('finish')
        return shuffle(ECG[1:],LABEL[1:],random_state=666) # 删除第一空行
    
    train = list(filter(flt,dataindexs))
    print(train)
    
    ECG_val,LABEL_val = load(val)
    ECG_test,LABEL_test = load(test)
    ECG_train,LABEL_train = load(train)

    return ECG_val,LABEL_val,ECG_test,LABEL_test,ECG_train,LABEL_train
    
# 不知道代码的底层逻辑，就无法写出高效的代码，头大
ECG_val,LABEL_val,ECG_test,LABEL_test,ECG_train,LABEL_train = split_train_val_test(val=[3,4],test=[5,6])
ECG_train.shape
# all_ = split_train_val_test(val=[1,2],test=[3,4])
# for i in all_:
#     print(i.shape)
```
