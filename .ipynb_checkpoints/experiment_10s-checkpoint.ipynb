```python
import time
import glob


import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.models import load_model
from keras.callbacks import ModelCheckpoint,EarlyStopping
from tensorflow.keras.layers import (
    Input,
    Conv1D,
    MaxPooling1D,
    Dropout,
    BatchNormalization,
    Activation,
    Add,
    Dense,
    GlobalAveragePooling1D,
)
import keras.backend as K
# from easydl import clear_output
from sklearn.utils import shuffle

# from get_representations import *

# 这次试验的思路是
# 第一步 数据处理 提取出ECG和对应标签，将5000降采样到4096，保存
# 第二步 定义test，val，train数据集的个体，对drink或者Exercise数据拼接成10s，即8192，
# 第三步 导入PCLR预训练模型，学习率0.001，mse，耐心50，epoch400，训练
# 第四步 使用BHS等验证模型效果
# ……
```

```python
dataindexs = [i for i in range(1,25)]
# dataindexs = [i for i in range(26,46,1)]

def split_train_val_test(val:list,test:list):
    """
    完成了每个数据集的读取和shuffle
    """
    def flt(x):
        return not (x in val or x in test)
    
    def load(indexs):
        ECG = np.empty((1,4096))
        LABEL = np.empty((1,2*2))
        for index in indexs:
            ecg = np.load('/data/drink/ECG/{}.npy'.format(index))
            label = np.load('/data/drink/label/label{}.npy'.format(index))

            ECG10 = []
            for i in range(int(ecg.shape[0]/2)):
                ECG10.append(np.concatenate([ecg[2*i],ecg[2*i+1]]))
                
            LABEL4 = []
            for i in range(int(label.shape[0]/2)):
                LABEL4.append(np.concatenate([label[2*i],label[2*i+1]]))
                
            ECG = np.vstack((ECG,ECG10))
            LABEL = np.vstack((LABEL,LABEL4))
            
#         降采样
#         ECG_ = np.zeros((ECG.shape[0],4096))
#         for i in range(ECG.shape[0]):
#             ECG_[i] = scipy.signal.resample_poly(ECG[i],4096,5000)
#         res = shuffle(ECG[1:],LABEL[1:],random_state=666)

        print('finish')
        return shuffle(ECG[1:],LABEL[1:],random_state=666) # 删除第一空行
    
    train = list(filter(flt,dataindexs))
    print(train)
    
    ECG_val,LABEL_val = load(val)
    ECG_test,LABEL_test = load(test)
    ECG_train,LABEL_train = load(train)

    return ECG_val,LABEL_val,ECG_test,LABEL_test,ECG_train,LABEL_train
    
# 不知道代码的底层逻辑，就无法写出高效的代码，头大
ECG_val,LABEL_val,ECG_test,LABEL_test,ECG_train,LABEL_train = split_train_val_test(val=[1,2],test=[])
ECG_train.shape
# all_ = split_train_val_test(val=[1,2],test=[3,4])
# for i in all_:
#     print(i.shape)
```

```python
def normalization(train,val,test):
    """对三个数据集统一归一化后，返回"""
    
    alldata = np.zeros((1,4096))
    alldata = np.vstack((alldata,train))
    alldata = np.vstack((alldata,val))
    alldata = np.vstack((alldata,test))
    
#     归一化
    ma = alldata.max()
    mi = alldata.min()
    alldata = (alldata-mi)/(ma-mi)
    
    print(ma)
    print(mi)
    s1 = train.shape[0]
    s2 = val.shape[0]
    s3 = test.shape[0]
     
    return alldata[1:s1+1],alldata[s1+1:s1+s2+1],alldata[s1+s2+1:]
```

```python
ECG_train,ECG_val,ECG_test = normalization(ECG_train,ECG_val,ECG_test)
ECG_train.shape
```

```python
pclr = load_model('/home/ubuntu/ml4h/ml4h/model_zoo/PCLR/zml/PCLR_lead_I.h5')

length = 32
model = tf.keras.models.Sequential()
model.add(pclr) # 记得改模型名称
model.add(Dense(length*10,activation='relu'))
model.add(Dense(length*5,activation='relu'))
model.add(Dense(length,activation='relu'))
model.add(Dense(4)) # sbp dbp
model.layers[0].trainable = False
model.summary()

# monitor:监视参数，min_delta:小于此数认为不变化，mode:loss小好，acc大好，patience:n周期无提升则退出，restore_best_weights:取最优权重
earlyStop = EarlyStopping(monitor='val_loss', 
                          min_delta=0, 
                          patience=50,
                          mode='min', 
                          verbose=1, 
                          restore_best_weights = True)

length = 32
epoch=300 # 前200个epoch不需要保存
batch_size=128
lr = 0.001
tp='mse'
opt = keras.optimizers.Adam(learning_rate=lr)
model.compile(optimizer=opt, loss=tp)

model_name = '/data/drink/e_result/1/320-160-32-4-{epoch:03d}-{val_loss:.2f}.h5'
checkpoint = ModelCheckpoint(filepath=model_name, 
                             monitor='val_loss', 
                             verbose=1, 
                             save_best_only=True, 
                             mode='min', 
                             save_best_value=True,
                             initial_value_threshold=200)

callback_list = [earlyStop, checkpoint]

h = model.fit(ECG_train,LABEL_train, 
              validation_data=(ECG_val,LABEL_val), 
              epochs=epoch, 
              batch_size=batch_size, 
              shuffle=True, 
              verbose=1,
              callbacks=callback_list
             )
```
