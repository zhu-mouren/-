```python
import time
import glob


import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import tensorflow as tf

# from easydl import clear_output
from sklearn.utils import shuffle

# from get_representations import *

def load_5sec(index:str):
    data_label = np.load('/data/drink/sub_{}.npz'.format(index))
    data = tf.transpose(data_label['arr_0'],(0,2,1))
    ECG = np.zeros((data.shape[0],5000)) # 分配空间，只需要ECG数据
    for index,line in enumerate(data):
        ECG[index] = line[0]
    return ECG,data_label['arr_1']

def normalization(data):
    """
    only for data
    data: 2d array 
    """
    print(data.shape)
    assert len(data.shape)==2
    
    for index,line in enumerate(data):
        max_ = line.max()
        min_ = line.min()
        data[index] = (line-min_)/(max_-min_)
    return data
```

```python
# 提取出ECG通道，5000下采样到4096，和对应的label
# for i in range(26,46,1):
for i in range(1,26,1):
    data,label = load_5sec(str(i))
    
    data_ = np.zeros((data.shape[0],2048))
    for x in range(data.shape[0]):
        data_[x] = scipy.signal.resample_poly(data[x],2048,5000)
    print(data_.shape)
#     plt.plot(data_)
#     plt.show()
    np.save('/data/drink/ECG/{}.npy'.format(str(i)),data_)
    np.save('/data/drink/label/label{}.npy'.format(str(i)),label)
    
# clear_output()
```
